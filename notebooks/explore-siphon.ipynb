```python
from siphon.catalog import TDSCatalog
from siphon.http_util import session_manager
```

```python
# Climate Data Gateway test
cat = TDSCatalog('https://tds.ucar.edu/thredds/catalog/catalog.xml')
print(list(cat.catalog_refs))
```

```python
rootref = cat.catalog_refs['NCAR Climate Data Gateway Root catalog']
rootcat = rootref.follow()

# Print all catalogs in the root catalog
allrootrefs = list(rootcat.catalog_refs)
print(allrootrefs)
```

```python
# Print Cordex-related refs in root catalog
cordexrefs = [ref for ref in allrootrefs if 'NAM' in ref]
print(cordexrefs)
```

```python
# Print Cordex-related refs in community catalog
commref = cat.catalog_refs['NCAR Climate Data Gateway Community catalog']
commcat = commref.follow()

# Print all catalogs in the root catalog
allrefs = list(commcat.catalog_refs)
cordexrefs = [ref for ref in allrefs if 'NAM' in ref]
print(cordexrefs)
```
